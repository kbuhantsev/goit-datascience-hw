```python
import pandas as pd
import numpy as np
```

Вивести перші рядки таблиці за допомогою методу head

```python
table_name = "Коефіцієнт народжуваності в регіонах України"
data_frame = pd.read_html("https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8", match=table_name)[0]

data_frame.head()
```

Визначте кількість рядків та стовпців у датафреймі (атрибут shape)

```python
data_frame.shape
```

Замініть у таблиці значення "—" на значення NaN

```python
data_frame = data_frame.replace(to_replace="—", value=np.NAN)

data_frame
```

Визначте типи всіх стовпців за допомогою dataframe.dtypes

```python
data_frame[data_frame.columns].dtypes
```

Замініть типи нечислових колонок на числові. Підказка - це колонки, де знаходився символ "—"

```python
data_frame[data_frame.columns[1:]] = data_frame[data_frame.columns[1:]].apply(pd.to_numeric, errors="coerce")

data_frame
```

Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)

```python
data_frame.isnull().sum()
```

Видаліть з таблиці дані по всій країні, останній рядок таблиці

```python
data_frame.drop(data_frame.tail(1).index, inplace=True)
data_frame
```

Замініть відсутні дані в стовпцях середніми значеннями цих стовпців (метод fillna)

```python
data_frame = data_frame.fillna(data_frame.mean(numeric_only=True))
data_frame
```
